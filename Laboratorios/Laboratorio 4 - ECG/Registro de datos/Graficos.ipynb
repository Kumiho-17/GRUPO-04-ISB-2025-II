## Tratamiento de seales


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch, find_peaks
# --- Funciones de filtrado ---
def butter_bandpass(lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a
def bandpass_filter(data, lowcut, highcut, fs, order=4):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = filtfilt(b, a, data)
    return y
def notch_filter(data, notch_freq, fs, quality_factor=30):
    nyq = 0.5 * fs
    w0 = notch_freq / nyq
    b, a = iirnotch(w0, quality_factor)
    y = filtfilt(b, a, data)
    return y
# --- Funciones para envolvente ---
def moving_average(signal, window_size):
    window = np.ones(window_size) / window_size
    return np.convolve(signal, window, mode='same')
def moving_rms(signal, window_size):
    squared = signal**2
    mean_squared = moving_average(squared, window_size)
    return np.sqrt(mean_squared)

fs = 1000  
lowcut = 0.5
highcut = 40.0
notch_freq = 60.0
nombres=["Reposo","Contencin de la respiracin - Toma 1","Recuperacin tras contencin - Toma 1","Contencin de la respiracin - Toma 2","Recuperacin tras contencin - Toma 2","Contencin de la respiracin - Toma 3","Recuperacin tras contencin - Toma 3","Actividad fsica","Despus de 15 minutos de actividad fsica"]
for i in range(len(l)):
    distancia_minima = int(0.5* fs) 
    senal_filtrada = bandpass_filter(l[i], lowcut, highcut, fs)
    senal_filtrada = notch_filter(senal_filtrada, notch_freq, fs)
# Graficar seal filtrada
    tiempo = np.arange(len(l[i])) / fs
    senal_rectificada = np.abs(senal_filtrada)
    window_size = int(0.150 * fs)  
    envolvente_media_movil = moving_average(senal_rectificada, window_size)
    envolvente_rms = moving_rms(senal_filtrada, window_size)

# Graficar envolvente (RMS)
    plt.figure(figsize=(15,4))
    plt.plot(tiempo, senal_filtrada, label='Seal Filtrada', alpha=0.5)
    plt.plot(tiempo, envolvente_rms, label='Envolvente (RMS)', color='green')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud (mV)')
    plt.title('Envolvente de la Seal ECG (RMS) - '+nombres[i])
    plt.legend()
    plt.grid()
    plt.show()
    altura_minima = np.max(senal_filtrada) * 0.5

    if i==2:
        distancia_minima = int(0.45*fs)  
        altura_minima = np.max(senal_filtrada) * 0.3
    elif i==4:
        distancia_minima = int(0.55*fs)  
        altura_minima = np.max(senal_filtrada) * 0.3
    elif i==6:
        distancia_minima = int(0.2*fs)  
        altura_minima = np.max(senal_filtrada) * 0.4
    elif i==7:
        distancia_minima = int(0.1*fs)  
        altura_minima = np.max(senal_filtrada) * 0.5

    picos_R, propiedades = find_peaks(senal_filtrada, distance=distancia_minima, height=altura_minima)
# Calcular intervalos R-R en segundos
    intervalos_RR = np.diff(picos_R) / fs
# Calcular frecuencia cardiaca instantnea (bpm)
    frecuencia_cardiaca = 60 / intervalos_RR
# Tiempos de los picos R
    tiempos_picos = picos_R / fs
# Graficar seal filtrada con picos R detectados
    plt.figure(figsize=(15,4))
    plt.plot(tiempo, senal_filtrada, label='Seal Filtrada')
    plt.plot(tiempos_picos, senal_filtrada[picos_R], 'ro', label='Picos R detectados')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Amplitud (mV)')
    plt.title('Deteccin de Picos R en Seal ECG - '+nombres[i])
    plt.legend()
    plt.grid()
    plt.show()
# Graficar frecuencia cardiaca instantnea
    plt.figure(figsize=(15,4))
    plt.plot(tiempos_picos[1:], frecuencia_cardiaca, marker='o')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Frecuencia Cardiaca (bpm)')
    plt.title('Frecuencia Cardiaca Instantnea (Serie R-R) - '+nombres[i])
    plt.grid()
    plt.show()




```
