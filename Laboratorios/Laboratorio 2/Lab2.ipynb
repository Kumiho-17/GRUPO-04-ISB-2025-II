### üìä Bases de datos
#### 1. NinaPro
<p align="justify">
NinaPro Database es una de la colecciones m√°s amplias utilizadas a nivel mundial para el estudio del reconocimiento de movimiento mioel√©ctricos de la mano y el control de la pr√≥tesis. Se cre√≥ debido a que se quer√≠a construir un benchmark p√∫blico y estandarizado para el desarrollo de m√©todos basado en clasificaci√≥n basados en la EMG [1].

**1.1. ‚öôÔ∏è Especificaciones T√©cnicas [1]:**
- Sujetos:
    - DB1: 27 sujetos sanos
    - DB2: 40 sujetos sanos
    - DB3: 11 sujetos amputados transradiales
- Movimientos registrados:
    - Entre 50 y 52 gestos de dedos, mano y mu√±eca, incluyendo movimientos b√°sicos, posturas isom√©tricas, y gestos funcionales de prensi√≥n.
- Electrodos y sensores:
    - DB1: 10 electrodos OttoBock colocados en el antebrazo.
    - DB2 y DB3: 12 electrodos Delsys Trigno inal√°mbricos (2 kHz) con aceler√≥metros integrados.
    - CyberGlove II (22 sensores articulares) y un inclin√≥metro 2D para capturar cinem√°tica de la mu√±eca.
- Protocolo de adquisici√≥n:
    - Cada gesto se repite 10 veces, con 5 s de contracci√≥n y 3 s de reposo.
    - Duraci√≥n total de una sesi√≥n ‚âà 100 minutos por sujeto.
- Formato de los datos: archivos ASCII sincronizados mediante timestamps; cada sujeto incluye se√±ales EMG, datos cinem√°ticos y est√≠mulos visuales.

**1.2. ‚úíÔ∏èUso en la literatura:**
- Primer reporte: base inicial con 27 sujetos, validaci√≥n con SVM (~80% en subsets reducidos) [2].
- Publicaci√≥n en TNSRE: se consolid√≥ como benchmark, con precisi√≥n m√°xima del 76% usando SVM no lineales [3].
- Publicaci√≥n oficial en Scientific Data (2014): ampliada a 78 sujetos (67 sanos y 11 amputados), validada con varios clasificadores, con resultados de ~75% en sanos y ~46% en amputados [4].
- Estudios recientes con deep learning alcanzan precisiones >95%.

**1.3. üîçRelevancia para el proyecto:**
<p align="justify">
NinaPro aporta una base valida y reconocida internacionalmente, lo que garantiza la comparabilidad de resultados y solidez metodol√≥gica . Su variedad de sujetos y gestos permite usar algoritmos b√°sicos hasta enfoques m√°s complejos.

#### 2. CapgMyo
<p align="justify">
El CapgMyo Database es un conjunto de datos de alta densidad de electromiograf√≠a de superficie (HD-sEMG) creado en la Universidad de Zhejiang (China). Fue dise√±ado para estudiar el reconocimiento de gestos mediante la representaci√≥n espacial de las se√±ales EMG, con un enfoque en interfaces m√∫sculo-computadora y control de pr√≥tesis [5].

**2.1. ‚öôÔ∏èEspecificaciones T√©cnicas [5]:**
- Sujetos: 18 j√≥venes sanos (edad 23‚Äì26 a√±os).
- Gestos registrados: 8 gestos isom√©tricos/isot√≥nicos de los dedos (id√©nticos a un subset de NinaPro).
- Electrodos: 128 canales distribuidos en 8 √ó 16 matrices (inter-electrodo 7.5 mm √ó 10.05 mm).
- Frecuencia de muestreo: 1,000 Hz, resoluci√≥n de 16 bits.
- Protocolo: cada gesto repetido 10 veces, con duraci√≥n de 3‚Äì10 s y 7 s de descanso entre repeticiones.
- Formato: se√±ales HD-sEMG convertidas en im√°genes de EMG (8 √ó 16 p√≠xeles por frame).

**2.2. ‚úíÔ∏èUso en la literatura [5]:**
<p align="justify">
El art√≠culo de presentaci√≥n de CapgMyo (Scientific Reports, Geng et al., 2016) introdujo el concepto de im√°genes instant√°neas de EMG para reconocimiento de gestos con deep learning (ConvNets):

- Con un solo frame de sEMG, alcanzaron 89.3% de precisi√≥n.
- Con votaci√≥n mayoritaria sobre 150 ms de se√±al, alcanzaron 99.5% de precisi√≥n en 8 gestos.
- Tambi√©n se prob√≥ el m√©todo en otras bases (NinaPro y CSL-HDEMG), mostrando que la t√©cnica supera m√©todos cl√°sicos (SVM, Random Forest, KNN, LDA).
- El estudio posicion√≥ a CapgMyo como un recurso clave para explorar representaciones espaciales de EMG y aplicaciones en control de pr√≥tesis y exoesqueletos.

**2.3. üîçRelevancia para el proyecto:**
<p align="justify">
El dataset CapgMyo se caracteriza por su alta densidad de se√±ales (128 canales), lo que permite analizar tanto la temporalidad como la distribuci√≥n espacial de la actividad EMG, favoreciendo la validaci√≥n de algoritmos avanzados. Su publicaci√≥n original lo consolid√≥ como una referencia en deep learning, al demostrar que las redes convolucionales (CNN) pueden alcanzar precisiones cercanas al 100% en el reconocimiento de gestos. Adem√°s, funciona como un complemento de NinaPro: mientras esta ofrece un marco cl√≠nico amplio con sujetos sanos y amputados, CapgMyo aporta datos de alta calidad experimental, √∫tiles para comparar m√©todos cl√°sicos frente a enfoques de aprendizaje profundo. Finalmente, la representaci√≥n de se√±ales en forma de ‚Äúim√°genes EMG‚Äù facilita su aplicaci√≥n educativa, introduciendo a los estudiantes al uso de redes convolucionales en biom√©dica mediante un caso pr√°ctico de reconocimiento de gestos.
