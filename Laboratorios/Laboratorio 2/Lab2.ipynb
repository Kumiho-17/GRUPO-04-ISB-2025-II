## :point_right:Proyecto: Reconocimiento de gestos de la mano mediante señales EMG de superficie: desarrollo de un sistema accesible con fines educativos y biomédicos
### :bar_chart:Bases de datos
#### 1. NinaPro
<p align="justify">
NinaPro Database es una de la colecciones más amplias utilizadas a nivel mundial para el estudio del reconocimiento de movimiento mioeléctricos de la mano y el control de la prótesis. Se creó debido a que se quería construir un benchmark público y estandarizado para el desarrollo de métodos basado en clasificación basados en la EMG [1].

**1.1. :gear:Especificaciones Técnicas [1]:**
- Sujetos:
    - DB1: 27 sujetos sanos
    - DB2: 40 sujetos sanos
    - DB3: 11 sujetos amputados transradiales
- Movimientos registrados:
    - Entre 50 y 52 gestos de dedos, mano y muñeca, incluyendo movimientos básicos, posturas isométricas, y gestos funcionales de prensión.
- Electrodos y sensores:
    - DB1: 10 electrodos OttoBock colocados en el antebrazo.
    - DB2 y DB3: 12 electrodos Delsys Trigno inalámbricos (2 kHz) con acelerómetros integrados.
    - CyberGlove II (22 sensores articulares) y un inclinómetro 2D para capturar cinemática de la muñeca.
- Protocolo de adquisición:
    - Cada gesto se repite 10 veces, con 5 s de contracción y 3 s de reposo.
    - Duración total de una sesión ≈ 100 minutos por sujeto.
- Formato de los datos: archivos ASCII sincronizados mediante timestamps; cada sujeto incluye señales EMG, datos cinemáticos y estímulos visuales.

**1.2. :black_nib:Uso en la literatura:**
- Primer reporte: base inicial con 27 sujetos, validación con SVM (~80% en subsets reducidos) [2].
- Publicación en TNSRE: se consolidó como benchmark, con precisión máxima del 76% usando SVM no lineales [3].
- Publicación oficial en Scientific Data (2014): ampliada a 78 sujetos (67 sanos y 11 amputados), validada con varios clasificadores, con resultados de ~75% en sanos y ~46% en amputados [4].
- Estudios recientes con deep learning alcanzan precisiones >95%.

**1.3. :mag:Relevancia para el proyecto:**
<p align="justify">
NinaPro aporta una base valida y reconocida internacionalmente, lo que garantiza la comparabilidad de resultados y solidez metodológica . Su variedad de sujetos y gestos permite usar algoritmos básicos hasta enfoques más complejos.

#### 2. CapgMyo
<p align="justify">
El CapgMyo Database es un conjunto de datos de alta densidad de electromiografía de superficie (HD-sEMG) creado en la Universidad de Zhejiang (China). Fue diseñado para estudiar el reconocimiento de gestos mediante la representación espacial de las señales EMG, con un enfoque en interfaces músculo-computadora y control de prótesis [5].

**2.1. :gear:Especificaciones Técnicas [5]:**
- Sujetos: 18 jóvenes sanos (edad 23–26 años).
- Gestos registrados: 8 gestos isométricos/isotónicos de los dedos (idénticos a un subset de NinaPro).
- Electrodos: 128 canales distribuidos en 8 × 16 matrices (inter-electrodo 7.5 mm × 10.05 mm).
- Frecuencia de muestreo: 1,000 Hz, resolución de 16 bits.
- Protocolo: cada gesto repetido 10 veces, con duración de 3–10 s y 7 s de descanso entre repeticiones.
- Formato: señales HD-sEMG convertidas en imágenes de EMG (8 × 16 píxeles por frame).

**2.2. :black_nib:Uso en la literatura [5]:**
<p align="justify">
El artículo de presentación de CapgMyo (Scientific Reports, Geng et al., 2016) introdujo el concepto de imágenes instantáneas de EMG para reconocimiento de gestos con deep learning (ConvNets):

- Con un solo frame de sEMG, alcanzaron 89.3% de precisión.
- Con votación mayoritaria sobre 150 ms de señal, alcanzaron 99.5% de precisión en 8 gestos.
- También se probó el método en otras bases (NinaPro y CSL-HDEMG), mostrando que la técnica supera métodos clásicos (SVM, Random Forest, KNN, LDA).
- El estudio posicionó a CapgMyo como un recurso clave para explorar representaciones espaciales de EMG y aplicaciones en control de prótesis y exoesqueletos.

**2.3. :mag:Relevancia para el proyecto:**
<p align="justify">
El dataset CapgMyo se caracteriza por su alta densidad de señales (128 canales), lo que permite analizar tanto la temporalidad como la distribución espacial de la actividad EMG, favoreciendo la validación de algoritmos avanzados. Su publicación original lo consolidó como una referencia en deep learning, al demostrar que las redes convolucionales (CNN) pueden alcanzar precisiones cercanas al 100% en el reconocimiento de gestos. Además, funciona como un complemento de NinaPro: mientras esta ofrece un marco clínico amplio con sujetos sanos y amputados, CapgMyo aporta datos de alta calidad experimental, útiles para comparar métodos clásicos frente a enfoques de aprendizaje profundo. Finalmente, la representación de señales en forma de “imágenes EMG” facilita su aplicación educativa, introduciendo a los estudiantes al uso de redes convolucionales en biomédica mediante un caso práctico de reconocimiento de gestos.

#### 3. BioPatrec
<p align="justify">
Se trata de una plataforma de investigación con datos en Matlab que permite implementar algoritmos de manera sencilla para ser probados en problemas de la vida real como el control de miembros artificiales mediante el reconocimiento de patrones de señales bioeléctricas [6].

**3.1. :gear:Especificaciones Técnicas [6]:**
- La plataforma incluye funcionalidades completas adquiridas mediante EMG, a su vez como un control en tiempo real mediante realidad virtual
- El estudio constó de 20 sujetos sanos
- Se contó con una frecuencia de muestreo de 2000 Hz en cuatro canales de EMG
- El muestreo fue en 121 ventanas de 200 ms.

**3.2. :black_nib:Uso en la literatura [6]:**
- El repositorio ha sido estudiado con el fin de observar su utilidad en la plataforma del NCBI para el reconocimiento de patrones mediante algoritmos.

**3.3. :mag:Relevancia para el proyecto:**
<p align="justify">
El repositorio es útil para el proyecto porque se alinea con los objetivos al contar con ejemplos de preprocesamiento, características, clasificación y evaluación del desempeño. Por otro lado, la base de datos permite probar sus señales mediante algoritmos personalizados y métricas que vienen con la propia base de datos. Finalmente, esta posee simuladores de realidad virtual con los que validar los resultados con respecto a las señales que se buscan evaluar.

#### 4. Kaggle
<p align="justify">
Kaggle es una plataforma de ciencia de datos creada en Melbourne, Australia, que funciona como un espacio colaborativo para científicos de todo el mundo. Ofrece miles de datasets, incluidos señales biomédicas como EMG, y competencias que permiten entrenar y comparar modelos predictivos. Es una herramienta clave para proyectos de reconocimiento de gestos y rehabilitación motora al facilitar el acceso a datos y recursos de análisis [7].

**4.1. :gear:Especificaciones Técnicas [8]:**
- Sujetos: 36 sujetos sanos
- Gestos registrados: 7 gestos estáticos básicos con las manos (mano en reposo, mano cerrada en puño, flexión de muñeca, extensión de muñeca, desviaciones radiales, desviaciones cubitales, palma extendida)
- Electrodos: Pulsera Tálmica MYO (8 electrodos)
- Frecuencia de muestreo: 200 Hz
- Protocolo: cada gesto se realiza durante 3 segundos con una pausa de 3 segundos entre gestos.
- Formato: Señales EMG

**4.2. :black_nib:Uso en la literatura [8]:**
<p align="justify">
El conjunto de datos “EMG Signal for Gesture Recognition” recopila señales EMG multicanales adquiridas con el brazalete Myo en 36 sujetos realizando gestos de la mano.

- Este recurso ha sido empleado en estudios de clasificación de gestos con machine learning tradicional, logrando precisiones superiores al 90% en gestos estáticos.

**4.3. :mag:Relevancia para el proyecto:**
<p align="justify">
El dataset de Kaggle ofrece señales EMG de sujetos que realizaron diferentes gestos de mano, siendo esto de suma relevancia para nuestro proyecto. Su acceso libre y estructurado permite validar los algoritmos de preprocesamiento y clasificación desarrollados, además de facilitar la comparación de resultados con los nuestros. Por último, es un recurso valioso en el ámbito educativo, ya que permite la replicación de experimentos en entornos de enseñanza biomédica al tener relación con los gestos aplicables en la rehabilitación motora.

#### 5. PhysioNet
<p align="justify">
PhysioNet es una plataforma creada por el MIT, que ofrece acceso gratuito a grandes bases de datos fisiológicos (ECG, EEG, EMG, respiración, presión arterial, entre otros), junto con herramientas de software para analizarlos. PhysioNet proporciona registros clínicos y experimentales que permiten estudiar patrones musculares en condiciones normales y patológicas. [9].

**5.1. :gear:Especificaciones Técnicas [10]:**
- Sujetos: 43 participantes sanos (edad: 24-35 años)
- Gestos registrados: 16 gestos con las manos y los dedos
- Electrodos: 28 electrodos monopolares de sEMG; antebrazo (16 electrodos) y muñeca (12 electrodos)
- Frecuencia de muestreo: 2048 Hz
- Protocolo: cada gesto se repitió siete veces (17 gestos x 7 ensayos = 119 contracciones), con cinco segundos de registro en cada ensayo y diez segundos de descanso entre contracciones
- Formato: Señales EMG


**5.2. :black_nib:Uso en la literatura [10]:**
<p align="justify">
El dataset GRABMyo, disponible en PhysioNet, recopila señales sEMG de 43 participantes sanos que realizaron 16 gestos manuales en distintas sesiones por tres días.

- Los datos han sido utilizados para probar algoritmos de clasificación en escenarios de rehabilitación y control de prótesis
- El estudio posiciona a GRABMyo como base de referencia para investigar sistemas de reconocimiento de gestos basados en EMG.


**5.3. :mag:Relevancia para el proyecto:**
<p align="justify">
El dataset en PhysioNet puede ser relevante para nuestro proyecto de reconocimiento de gestos de la mano mediante EMG, porque proporciona un conjunto amplio y estandarizado de señales con diferentes gestos, sujetos y sesiones. Esto permite entrenar algoritmos de clasificación en condiciones realistas. Además, al ser una página abierta se convierte en un gran recurso para el desarrollo de un sistema accesible para los alumnos, cayendo en línea con los objetivos de nuestro proyecto.

### :thinking:¿Qué base de datos usaremos?
<p align="justify">
Después de comparar las cinco bases de datos analizadas, la más adecuada para este proyecto es NinaPro, ya que constituye el estándar internacional en el reconocimiento de gestos mioeléctricos. Su amplitud en número de sujetos, variedad de gestos y validación en más de 200 publicaciones garantizan la solidez metodológica y la comparabilidad de resultados. Como complemento, CapgMyo resulta especialmente útil para explorar enfoques avanzados como el deep learning y el análisis espacial de señales de alta densidad, enriqueciendo la validación experimental del proyecto. Las demás bases de datos (BioPatRec, Kaggle y PhysioNet) son valiosas en contextos educativos o clínicos específicos, pero no ofrecen la misma cobertura ni relevancia para los objetivos principales de este trabajo.
