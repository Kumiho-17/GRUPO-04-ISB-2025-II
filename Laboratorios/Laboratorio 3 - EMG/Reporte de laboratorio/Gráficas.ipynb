```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch
Vref = 3.3  # Voltaje de referencia en V
ADC_max = 1023  # Para 10 bits
señales=[]
movimientos=["Reposo - Bíceps braquial","Movimiento libre - Bíceps braquial","Movimiento en contra - Bíceps braquial","Reposo - Flexor carpi radialis","Movimiento libre - Flexor carpi radialis","Movimiento en contra - Flexor carpi radialis","Reposo - Abductor pollicis brevis","Movimiento libre - Abductor pollicis brevis","Movimiento en contra - Abductor pollicis brevis"]
fs = 1000  

prom_R_P1 = np.mean(R_P1, axis=0)
señales.append(prom_R_P1)
prom_ML_P1 = np.mean(ML_P1, axis=0)
señales.append(prom_ML_P1)
prom_MC_P1 = np.mean(MC_P1, axis=0)
señales.append(prom_MC_P1)
prom_R_P2 = np.mean(R_P2, axis=0)
señales.append(prom_R_P2)
prom_ML_P2 = np.mean(ML_P2, axis=0)
señales.append(prom_ML_P2)
prom_MC_P2 = np.mean(MC_P2, axis=0)
señales.append(prom_MC_P2)
prom_R_P3 = np.mean(R_P3, axis=0)
señales.append(prom_R_P3)
prom_ML_P3 = np.mean(ML_P3, axis=0)
señales.append(prom_ML_P3)
prom_MC_P3 = np.mean(MC_P3, axis=0)
señales.append(prom_MC_P3)
m=0
for i, señal in enumerate(señales):

    señal= (señal/ ADC_max) * Vref * 1000
    
# 1. Eliminar DC 
    señal_sin_dc = señal - np.mean(señal)

# 2. Filtro pasa banda 20-450 Hz
    def bandpass_filter(data, fs, low=20, high=450, order=4):
        nyq = 0.5 * fs
        b, a = butter(order, [low/nyq, high/nyq], btype='band')
        return filtfilt(b, a, data)

    señal_bandpass = bandpass_filter(señal_sin_dc, fs)

# 3. Filtro notch 
    def notch_filter(data, fs, freq=60, Q=30):
        b, a = iirnotch(freq/(fs/2), Q)
        return filtfilt(b, a, data)

    señal_preprocesada = notch_filter(señal_bandpass, fs)

# 4. Rectificación + envolvente (RMS con ventana de 100 ms)
    ventana = int(0.1 * fs)  # 100 ms
    señal_rectificada = np.abs(señal_preprocesada)
    envolvente_rms = np.sqrt(np.convolve(señal_rectificada**2, np.ones(ventana)/ventana, mode='same'))

    t = np.arange(len(señal)) / fs

    fig, axs = plt.subplots(1, 3, figsize=(18, 5), sharex=True)
    fig.suptitle(movimientos[m], fontsize=16)
    m+=1
    axs[0].plot(t, señal, color='blue')
    axs[0].set_title('Promedio original')
    axs[0].set_ylabel('Amplitud (mV)')
    axs[0].grid()

    axs[1].plot(t, señal_preprocesada, color='orange')
    axs[1].set_title('Preprocesada (sin DC + band-pass + notch)')
    axs[1].set_ylabel('Amplitud (mV)')
    axs[1].grid()

    axs[2].plot(t, envolvente_rms, color='green')
    axs[2].set_title('Rectificada + Envolvente RMS')
    axs[2].set_ylabel('Amplitud (mV)')
    axs[2].set_xlabel('Tiempo (s)')
    axs[2].grid()

    plt.tight_layout()
    plt.show()

    



```
