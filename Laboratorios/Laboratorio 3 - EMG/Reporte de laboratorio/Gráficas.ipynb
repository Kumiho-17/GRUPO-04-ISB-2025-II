```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, iirnotch
from pathlib import Path

# =========================
# Parámetros de adquisición
# =========================
fs        = 1000        # Hz (verifica tu frecuencia real)
canal     = 5           # columna (0-index) donde está el EMG
Vcc       = 3.3         # Voltaje de alimentación (V)
GAIN_EMG  = 1009        # Ganancia aproximada del canal EMG (BITalino EMG ~1009)
ADC_BITS  = 10
ADC_MAX   = (2**ADC_BITS) - 1

# =========================
# Parámetros de procesado
# =========================
ventana_rms_s = 0.10    # 100 ms
bp_low, bp_high = 20, 450  # Hz
notch_freqs = [60, 120] # Perú (red eléctrica y 1er armónico)
Q_notch = 30
bp_order = 4

# =========================
# Conversión ADC -> mV
# =========================
def adc_to_mV(counts, Vcc=Vcc, gain=GAIN_EMG, adc_max=ADC_MAX, offset=None):
    """
    Convierte cuentas ADC (centradas en Vcc/2) a milivoltios (mV) en la ENTRADA EMG.
    """
    x = np.asarray(counts, dtype=float)
    if offset is None:
        # Usar la mediana del segmento (o mejor: mediana de una ventana de reposo del mismo archivo)
        offset = np.median(x)
    V_adc = (x - offset) * (Vcc / adc_max)  # Voltios a la salida del ADC, centrados en 0
    V_emg = V_adc / gain                    # Voltios a la entrada EMG (antes de la ganancia)
    return 1e3 * V_emg                      # mV

# =========================
# Filtros y features
# =========================
def bandpass(x, fs, low=20, high=450, order=bp_order):
    nyq = 0.5*fs
    b, a = butter(order, [low/nyq, high/nyq], btype='band')
    return filtfilt(b, a, x)

def notch(x, fs, f0=60, Q=Q_notch):
    b, a = iirnotch(f0/(fs/2), Q)
    return filtfilt(b, a, x)

def preprocess(x):
    """
    Quita DC (centrado por media), aplica BP 20–450 Hz y notch 60/120 Hz.
    (x ya viene en mV)
    """
    x = x - np.mean(x)                # quitar DC residual
    x = bandpass(x, fs, bp_low, bp_high)
    for f0 in notch_freqs:
        x = notch(x, fs, f0, Q_notch)
    return x

def rms_envelope(x, fs, win_s=ventana_rms_s):
    N = max(1, int(win_s*fs))
    x2 = x**2
    kern = np.ones(N)/N
    env = np.sqrt(np.convolve(x2, kern, mode='same'))
    return env

# =========================
# Carga de segmentos (ya en mV)
# =========================
def load_seg_mV(path, start, end, ch=canal, offset=None):
    """
    Carga el archivo, extrae [start:end] del canal ch y lo convierte a mV.
    Si conoces una ventana de reposo en el MISMO archivo, pásala como offset
    (o calcula offset fuera y pásalo como número).
    """
    arr = np.loadtxt(path, dtype=float)
    seg_counts = arr[start:end, ch]
    seg_mV = adc_to_mV(seg_counts, offset=offset)  # <-- conversión aquí
    return seg_mV

# =========================
# QC (simple) en mV
# =========================
def qc_clip_mV(x_mV):
    """
    Descarta si >=0.5% de puntos superan 5*DE (heurística simple).
    
    """
    thr = 5*np.std(x_mV)
    frac = np.mean(np.abs(x_mV) >= thr)
    return frac < 0.005

def stack_mean_equal_length(list_of_1d):
    """
    Recorta todas al largo mínimo y promedia.
    """
    L = min(len(v) for v in list_of_1d)
    if L == 0:
        raise ValueError("Segmento vacío.")
    M = np.vstack([v[:L] for v in list_of_1d])
    return M.mean(axis=0)

# =========================
# Plot
# =========================
def plot_triptych(t, x_raw_mV, x_pre_mV, env_mV, titulo):
    fig, ax = plt.subplots(1,3, figsize=(18,4), sharex=True)
    fig.suptitle(titulo, fontsize=14)

    ax[0].plot(t, x_raw_mV, color="purple");            ax[0].set_title("Cruda")
    ax[0].set_ylabel("Amplitud (mV)")
    ax[1].plot(t, x_pre_mV, color='purple'); ax[1].set_title("Preprocesada")
    ax[1].set_ylabel("Amplitud (mV)")
    ax[2].plot(t, env_mV,  color='purple');  ax[2].set_title("Rectificada + Envolvente RMS")
    for a in ax:
        a.grid(True); a.set_xlabel("Tiempo (s)")
    plt.tight_layout(); plt.show()

# =========================
# Archivos por músculo/fase
# =========================
base = Path(r"C:\Users\RODRIGO\Desktop\GRUPO-04-ISB-2025-II\Laboratorios\Laboratorio 3 - EMG\Registro de datos")

files = {
    "Bíceps braquial": {
        "Reposo": [
            (base/"Posición 1/Toma1_Reposo_Posición 1.txt", 50000, 60000),
            (base/"Posición 1/Toma2_Reposo_Posición 1.txt",     0, 10000),
        ],
        "Movimiento": [
            (base/"Posición 1/Toma1_Movimiento libre_Posición1.txt",  7000, 17000),
            (base/"Posición 1/Toma2_Movimiento libre_Posición1.txt", 12000, 22000),
            (base/"Posición 1/Toma3_Movimiento libre_Posición1.txt",  7000, 17000),
        ],
        "Resistencia": [
            (base/"Posición 1/Toma1_Movimiento en contra_Posición1.txt", 5000, 15000),
            (base/"Posición 1/Toma2_Movimiento en contra_Posición1.txt", 4000, 14000),
            (base/"Posición 1/Toma3_Movimiento en contra_Posición1.txt", 7500, 17500),
        ],
    },
    "Flexor carpi radialis": {
        "Reposo": [
            (base/"Posición 2/Toma1_Reposo_Posición2.txt", 10000, 20000),
            (base/"Posición 2/Toma2_Reposo_Posición2.txt",     0, 10000),
        ],
        "Movimiento": [
            (base/"Posición 2/Toma1_Movimiento libre_Posición2.txt", 0, 10000),
            (base/"Posición 2/Toma2_Movimiento libre_Posición2.txt", 0, 10000),
            (base/"Posición 2/Toma3_Movimiento libre_Posición2.txt", 0, 10000),
        ],
        "Resistencia": [
            (base/"Posición 2/Toma1_Movimiento en contra_Posición2.txt", 0, 10000),
            (base/"Posición 2/Toma2_Movimiento en contra_Posición2.txt", 6000, 16000),
            (base/"Posición 2/Toma3_Movimiento en contra_Posición2.txt", 4000, 14000),
        ],
    },
    "Abductor pollicis brevis": {
        "Reposo": [
            (base/"Posición 3/Toma1_Reposo_Posición3.txt", 0, 7000),
            (base/"Posición 3/Toma2_Reposo_Posición3.txt", 0, 7000),
        ],
        "Movimiento": [
            (base/"Posición 3/Toma1_Movimiento libre_Posición3.txt", 0, 10000),
            (base/"Posición 3/Toma2_Movimiento libre_Posición3.txt", 0, 10000),
            (base/"Posición 3/Toma3_Movimiento libre_Posición3.txt", 0, 10000),
        ],
        "Resistencia": [
            (base/"Posición 3/Toma1_Movimiento en contra_Posición3.txt", 0, 8000),
            (base/"Posición 3/Toma2_Movimiento en contra_Posición3.txt", 0, 8000),
            (base/"Posición 3/Toma3_Movimiento en contra_Posición3.txt", 0, 8000),
        ],
    }
}

# =========================
# Pipeline por músculo/fase con promedio (en mV)
# =========================
for musc, fases in files.items():
    for fase, tomas in fases.items():
        segmentos_crudos_mV = []
        for ruta, i0, i1 in tomas:
            try:
                x_mV = load_seg_mV(ruta, i0, i1, ch=canal)
            except Exception as e:
                print(f"[ERR] No se pudo leer {ruta}: {e}")
                continue

            if qc_clip_mV(x_mV):   # QC simple
                segmentos_crudos_mV.append(x_mV)
            else:
                print(f"[QC] Excluida por clipping: {ruta.name}")

        if len(segmentos_crudos_mV) == 0:
            print(f"[WARN] Sin tomas válidas para {musc} - {fase}")
            continue

        # Promedio (asegurando misma longitud)
        x_raw_mean_mV = stack_mean_equal_length(segmentos_crudos_mV)   # mV
        # Preprocesado
        x_pre_mV      = preprocess(x_raw_mean_mV)                      # mV
        # Envolvente
        env_mV        = rms_envelope(x_pre_mV, fs, ventana_rms_s)      # mV

        t = np.arange(len(x_raw_mean_mV)) / fs
        plot_triptych(t, x_raw_mean_mV, x_pre_mV, env_mV, f"{fase} - {musc}")



```
